```python
import pandas as pd 
import numpy as np 
import os 
```

```python
purchase_data = os.path.join('.', 'purchase_data.csv')
```

```python
#create data frame
df = pd.read_csv(purchase_data)
pd.DataFrame.drop_duplicates(df)
df.head()
```

```python
# Player Count (Total number of players)
uniqueSN = df["SN"].unique()
unique_players = len(uniqueSN)

total_players_dict = {
    "Total Players": [unique_players]
}
total_players_df = pd.DataFrame(total_players_dict)
total_players_df
```

```python
#Purchasing Analysis
#Number of unique items
uniqueID = df["Item ID"].unique()
item_count = len(uniqueID)
#Average purchase price
average_purchase_price = df["Price"].mean()
#Number of purchases
purchase_count = df["Price"].count()
#Total Revenue
total_revenue = df["Price"].sum()
#Create Data frame
summary_table = pd.DataFrame({"Number of Unique Items": [item_count],
                              "Average Purchase Price": [average_purchase_price],
                              "Number of Purchase": [purchase_count],
                              "Total Revenue": [total_revenue]})
summary_table
```

```python
#Gender Demographics
males = df.loc[df["Gender"] == "Male",:]
females = df.loc[df["Gender"] == "Female",:]

others_list = []
for x in range(0, len(df["Gender"])):
    if(df["Gender"][x] != "Male" and df["Gender"][x] != "Female"):
        others_list.append(df["Gender"][x])

others_set = set(others_list)
other = df.loc[df["Gender"] == others_list[0],:]

player_gender_counts_list = []
gender_percentages_list = []
male_players = len(males["SN"].unique())
player_gender_counts_list.append(male_players)
female_players = len(females["SN"].unique())
player_gender_counts_list.append(female_players)
other_players = len(other["SN"].unique())
player_gender_counts_list.append(other_players)


for d in range(0, len(player_gender_counts_list)):
    gender_percentages_list.append(player_gender_counts_list[d]/unique_players)

gend_demographics_dict = {"Gender": ["Male", "Female", "Other / non-Disclosed"],
                          "Number of Players": player_gender_counts_list,
                          "Percentage": gender_percentages_list}
gender_demographics = pd.DataFrame(gend_demographics_dict)
gender_demographics
```

```python
#Purchasing Analysis (Gender)
#Purchase count
purchase_count_list = []
purchase_count_list.append(len(males["Price"]))
purchase_count_list.append(len(females["Price"]))
purchase_count_list.append(len(other["Price"]))
#Average purchase price and total puchase price
average_list = []
total_list = []
total_list.append(sum(males["Price"]))
total_list.append(sum(females["Price"]))
total_list.append(sum(other["Price"]))
for x in range(0, len(total_list)):
    average_list.append(total_list[x]/purchase_count_list[x])
    
gender_dict = {"Gender": ["Male", "Female", "Other / Non-Disclosed"],
               "Purchase Count": purchase_count_list,
               "Average Purchase Price": average_list,
               "Total Purchase Value": total_list}
gender_df = pd.DataFrame(gender_dict, columns=["Gender", "Purchase Count", "Average Purchase Price", "Total Purchase Value"])
gender_df
```

```python
#Age Demographics
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
df["Age Ranges"] = pd.cut(df["Age"], bins, labels=labels)
age_demo_totals = df["Age Ranges"].value_counts()
age_demo_percents = age_demo_totals / unique_players * 100
age_demo = pd.DataFrame({"Total Count": age_demo_totals, "Percent of Players": age_demo_percents})
age_demo = age_demo.sort_index()
age_demo
```

```python
#Purchasing Analysis (Age)
df["Age Ranges"] = pd.cut(df["Age"], bins, labels=labels)
avg_purchase_price = df.groupby(["Age Ranges"]).mean()["Price"]
total_purchase_value = df.groupby(["Age Ranges"]).sum()["Price"]
age_count = df.groupby(["Age Ranges"]).count()["Price"]
age_data = pd.DataFrame({"Purchase Count": age_count,
                         "Average Purchase Price": avg_purchase_price,
                         "Total Purchase Value": total_purchase_value})
age_data
```

```python
top_purchase_values = df.groupby(["SN"]).sum()["Price"]
top_average_price = df.groupby(["SN"]).mean()["Price"]
top_purchase_count = df.groupby(["SN"]).count()["Price"]
top_data = pd.DataFrame({"Purchase Count": top_purchase_count,
                         "Average Purchase Price": top_average_price,
                         "Total Purchase Value": top_purchase_values})
summary_table_top_spenders = top_data.sort_values("Total Purchase Value", ascending=False).head(5)
summary_table_top_spenders
```

```python
#Most popular items
item_data = df.loc[:,["Item ID", "Item Name", "Price"]]
item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"]
item_price = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_item_value = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"]

item_data_pd = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_price,
                             "Total Purchase Value": total_item_value})
item_data_pd = item_data_pd.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]
popular_items = item_data_pd.sort_values("Purchase Count", ascending=False).head(5)
summary_popular_items = popular_items.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]
summary_popular_items
```
